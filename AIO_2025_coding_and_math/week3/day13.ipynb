```python
import numpy as np
```

```python
A = np . array ([[1 , 2] , [3 , 4]])
B = np . array ([[5 , 6] , [7 , 8]])

C = A + B # Cộng ma trận
D = A - B # Trừ ma trận
```

```python
C
```

```python
D
```

```python
E = np.dot(A,B)
E
```

```python
det_A = np . linalg . det ( A )
```

```python
det_A
```

```python
A_inv = np . linalg . inv ( A )
A_inv
```

```python
np.dot(A,A_inv)
```

```python
rank_A = np . linalg . matrix_rank ( A )
rank_A
```

```python
x = np . linalg . solve (A , B )
x
```

```python
eig_values , eig_vectors = np . linalg . eig ( A )
print(eig_values ,'\n', eig_vectors)
```

```python
U , S , V = np . linalg . svd ( A )
print(U , S , V,sep = "\n------------\n")
```
