```python
import math
```

```python
def create_position_matrix ( seq_length , embed_size ) :
    position_matrix = [
        [0 for _ in range ( embed_size ) ]
        for _ in range (seq_length ) 
        ]
    for pos in range(seq_length):
        for i_emb in range(embed_size):
            if i_emb%2==0:
                position_matrix[pos][i_emb] = math.sin(
                    pos / (10000 ** (2*i_emb/embed_size))
                )
            else:
                position_matrix[pos][i_emb] = math.cos(
                    pos/(10000**(2*i_emb/embed_size))
                )
    
    # Your Code Here #
    return position_matrix

# Test
seq_length = 10
embed_size = 16
position_matrix = create_position_matrix ( seq_length , embed_size )
position_matrix
```
